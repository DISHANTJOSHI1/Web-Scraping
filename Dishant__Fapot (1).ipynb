```python
from lxml import html
import requests
from bs4 import BeautifulSoup, SoupStrainer
import pandas as pd

headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

page = requests.get('https://www.agtta.co.in/individuals.php',headers=headers)
tree = html.fromstring(page.content)
#This will create a list of MobioleNumbers:
MobileNumbers = tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/../p/text()')
#This will create a list of Names
Names = tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/text()')
Company = tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/..//div[2]/text()')
Address = tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/..//div[3]/text()')

Email = tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/..//div[4]/text()')

PhoneNumbers = tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/..//div[5]/text()')
FaxNumber = tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/..//div[6]/text()')
Route=tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/..//div[6]/text()')
Route01=tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/..//div[7]/text()')
Route02=tree.xpath('//h3[contains(@class,"b-branches__title ui-title-inner ui-title-inner_lg")]/..//div[4]/text()')


myDict = {}
for (i,j,k,l,m,n,o,p,q,r) in zip(Names,MobileNumbers,Company,Address,Email,PhoneNumbers,FaxNumber,Route,Route01,Route02):
    myDict[i]=[j,k,l,m,n,o,p,q,r]


#print(myDict.items())
df=pd.DataFrame(myDict.items(), columns=['Name', 'All'])
df
df[['MobileNumbers','Company','Address','Email','PhonNumbers','FaxNumber','Route','Route01','Route02']] = pd.DataFrame(df.All.tolist(), index= df.index)
#df['FaxNumber'] = df['FaxNumber'].str.extract('(^[0-9]*-[0-9]*)', expand=False)
#df['FaxNumber01'] = df['FaxNumber'].str.extract('(\^[0-9]*-[0-9]{7})', expand=True)
df['FaxNumber'] = df['FaxNumber'].str.replace(r"^[a-zA-Z].*",'')
df['FaxNumber'] = df['FaxNumber'].str.replace(r"^[0-9]*-*[0-9]*-[0-9]*",'')
df['FaxNumber'] = df['FaxNumber'].str.replace(r".*[(?].*",'')

df['PhonNumbers'] = df['PhonNumbers'].str.replace(r"^[0-9]*-*[0-9]*-[0-9]*",'')
df['PhonNumbers'] = df['PhonNumbers'].str.replace(r"[a-zA-Z \s&]*",'')
df['PhonNumbers'] = df['PhonNumbers'].str.replace(r",,,",'')
df['PhonNumbers'] = df['PhonNumbers'].str.replace(r",/",'')

df['Route01'] = df['Route01'].str.replace(r"^[0-9]*-*[0-9]*-[0-9]*",'')
df['Route01'] = df['Route01'].str.replace(r".*[0-9]*-[0-9]*$",'')
df['Route02'] = df['Route02'].str.replace(r'^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$','')

df['Route'] = df['Route'].str.replace(r'.*[0-9]*-[0-9]*$','')


df['Route02'] = df['Route02'].str.replace(r'[a-zA-Z0-9_.]*@[a-zA-Z]*.?[a-zA-Z].*','')




#df['Route'] = df['Route'].str.replace(r"^[0-9].*",'')
df['Email'] = df['Email'][df['Email'].str.contains(r'[a-zA-Z0-9]*@[a-zA-Z]*.?[a-zA-Z]*')]
df
#extract('(^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$)',expand=False)

df.to_csv('C:\\Users\\Education\\Desktop\\Dishant.csv', index=False)

#df[['team1','team2','a','d']] = pd.DataFrame(df.All.tolist(), index= df.index)
#df
#df['All']
#df[['MobileNumber01','MobileNumber02','A']] = df['All'].str.split(",",expand=True)
```

```python
import requests
import csv
from bs4 import BeautifulSoup as soup
import re

url = 'https://www.agtta.co.in/individuals.php'

csv_writer = csv.writer(open('scrape.csv', 'w'))

def remove_tags(text):
    TAG_RE = re.compile(r'<[^>]+>')
    return TAG_RE.sub('', text)

page =requests.get(url,headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:81.0) Gecko/20100101 Firefox/81.0"})
soups = soup(page.content.decode('utf-8'), 'html.parser')
name = soups.findAll('h3', {'class': "b-branches__title ui-title-inner ui-title-inner_lg"})
numbers_memebers = soups.findAll('p')
individual_reports = soups.findAll('div', {'class' : 'b-branches__item'})

fields = ["Name", "Mobile No", "Mobile No2", "Company Name", "Address", "Email", "Phone Number1", "Phone Number2", "Fax", "Route", "B Date", "J Date"]
data = []
namess = []
number = []
report = []
individuals = []
for names in name:
    namess.append(remove_tags(str(names)))

for numbers in numbers_memebers:
    n = remove_tags(str(numbers))
    lists = n.split()
    number.append(''.join([i for i in lists if i not in "Mobile No"]))
    for i in number:
        if "LIFETIME" in i:
            number.remove(i)

for reports in individual_reports:
    r = remove_tags(str(reports))  
    infos = r.split()
    report.append(' '.join([j for j in infos if j not in ("Firm","Address", "Email", "Phone No", "Fax No", "Route", "B. Date", "J. Date")]))
  
i=0
while i<len(report):
  individuals.append(report[i:i+7])
  i+=7

print(individuals)

for k in range(159):
    if "," in number[k]:
        number2 = ",".join(number).split(",")
        data.append([namess[k], number2[k], number2[k+1]])
    else:
        data.append([namess[k], number[k], ""])

csv_writer.writerow(fields)
csv_writer.writerows(data)
```
